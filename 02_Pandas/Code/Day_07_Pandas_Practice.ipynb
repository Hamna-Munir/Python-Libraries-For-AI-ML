## Groupby on IPL Dataset

```python
delivery = pd.read_csv('deliveries.csv')
```

```python
delivery.head(3)
```

```python
runs =delivery.groupby('batsman')
```

```python
runs['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
mask=delivery['batsman_runs']==6
new_delivery=delivery[mask]
```

```python
new_delivery.shape[0]
```

```python
new_delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head()
```

```python
vk=delivery[delivery['batsman']=='V Kohli']
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
def run_scored(batsman_name):
    vk=delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
run_scored('DA Warner')
```

```python
run_scored('G Gambhir')
```

## The isin() function

```python
death_over=delivery[delivery['over']>15]
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
all_batsman[x]
```

```python
all_batsman[x].shape
```

```python
batsman_list=all_batsman[x].index.tolist()
```

```python
#balls played by these 43 batsman
final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
final.groupby('batsman')['batsman'].sum()
```

## The Merge Function

```python
match=pd.read_csv('matches.csv')
```

```python
delivery.merge(match,left_on='match_id',right_on='id').shape
```

```python
print(delivery.shape)
print(match.shape)
```

```python
new=delivery.merge(match,left_on='match_id',right_on='id')
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum()
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first')[['season','batsman']]
```
