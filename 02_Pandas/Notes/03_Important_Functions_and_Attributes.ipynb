This notebook covers the most frequently used Pandas functions and essential DataFrame/Series attributes that are vital for data analysis.

## 1. Importing Pandas

```
import pandas as pd
print('Pandas Version:', pd.__version__)
```

## 2. Loading a Sample Dataset

```
df = pd.DataFrame({
    'Name': ['Ali', 'Sara', 'Raza', 'Hamna'],
    'Age': [23, 21, 25, 22],
    'Marks': [85, 92, 78, 90]
})
df
```

## 3. Important Attributes

```
print('Shape:', df.shape)
print('Columns:', df.columns)
print('Index:', df.index)
print('Data Types:\n', df.dtypes)
```

## 4. Important Functions

```
# Head and Tail
display(df.head())
display(df.tail())
```

```
# Statistical Summary
df.describe()
```

```
# Info Summary
df.info()
```

## 5. Accessing Columns and Rows
Selecting specific columns using `df['col']` or `df.col`.

```
df['Name']
```

```
df[['Name', 'Marks']]
```

### Selecting Rows using loc and iloc

```
df.loc[0]   # Select row by label
```

```
df.iloc[1]  # Select row by position
```

## 6. Checking for Missing Values

```
df.isnull()
```

```
df.isnull().sum()
```
