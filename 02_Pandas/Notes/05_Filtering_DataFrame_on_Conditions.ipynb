Filtering DataFrames in Pandas allows you to **select rows that satisfy specific conditions**. This is crucial for **data analysis, cleaning, and preprocessing** before AI/ML modeling.

---

## ðŸ“˜ Why Filter Data?
- To focus on relevant data.
- To remove unwanted rows.
- To perform analysis on a subset of data.
- To combine multiple conditions for complex queries.

## ----------------------------------------------------------
## Importing Pandas and Creating Sample DataFrame
## ----------------------------------------------------------
We will use a sample DataFrame to demonstrate filtering.

```python
import pandas as pd

data = {
    'Name': ['Ali', 'Sara', 'Umar', 'Zoya', 'Omar'],
    'Age': [22, 25, 28, 30, 26],
    'Salary': [50000, 60000, 70000, 80000, 65000],
    'City': ['NY', 'LA', 'Chicago', 'Boston', 'NY']
}
df = pd.DataFrame(data)
print("Sample DataFrame:\n", df)
```

## ðŸ§© Filtering with a Single Condition
You can filter rows where a column meets a specific condition using **boolean expressions**.

### Example: Select rows where Age > 25

```python
# Filter rows where Age > 25
filtered_age = df[df['Age'] > 25]
print("Rows with Age > 25:\n", filtered_age)
```

## ðŸ§© Filtering with Multiple Conditions
Use `&` (AND) and `|` (OR) operators to combine conditions.
Always wrap each condition in parentheses `()`.

### Example: Age > 25 AND City is 'NY'

```python
# Multiple conditions
filtered_multi = df[(df['Age'] > 25) & (df['City'] == 'NY')]
print("Rows with Age > 25 AND City = NY:\n", filtered_multi)
```

### Example: Age > 25 OR City is 'LA'

```python
# OR condition
filtered_or = df[(df['Age'] > 25) | (df['City'] == 'LA')]
print("Rows with Age > 25 OR City = LA:\n", filtered_or)
```

## ðŸ§© Filtering with `isin()`
- Use `isin()` to filter rows where a column value is in a list.

### Example: Select rows where City is either 'NY' or 'LA'

```python
# Using isin()
filtered_isin = df[df['City'].isin(['NY', 'LA'])]
print("Rows where City is NY or LA:\n", filtered_isin)
```

## ðŸ§© Filtering with `~` (NOT) Operator
- `~` negates a condition.

### Example: Select rows where City is NOT 'NY'

```python
# NOT condition
filtered_not = df[~df['City'].isin(['NY'])]
print("Rows where City is NOT NY:\n", filtered_not)
```

## ðŸ“ Summary
- Filtering DataFrames allows **subset selection** based on **conditions**.
- Single conditions use standard comparison operators: `>`, `<`, `==`, `!=`, `>=`, `<=`.
- Multiple conditions can be combined using `&` (AND) and `|` (OR).
- `isin()` helps filter rows based on a list of values.
- `~` negates conditions for excluding data.
