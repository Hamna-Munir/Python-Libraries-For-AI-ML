## ðŸ”¹ 1. Linear Regression
Linear regression fits a straight line that best describes the relationship between variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
tips = sns.load_dataset('tips')
sns.regplot(data=tips, x='total_bill', y='tip')
plt.title('Linear Regression: Total Bill vs Tip')
plt.show()
```

## ðŸ”¹ 2. Polynomial Regression
Polynomial regression models a curved trend using the `order` parameter.

```python
sns.regplot(data=tips, x='total_bill', y='tip', order=2)
plt.title('Polynomial Regression (Order 2)')
plt.show()
```

## ðŸ”¹ 3. Confidence Intervals
Seaborn adds confidence intervals automatically. Adjust using `ci` or `errorbar`.

```python
sns.regplot(data=tips, x='total_bill', y='tip', ci=95)
plt.title('Regression with 95% Confidence Interval')
plt.show()
```

## ðŸ”¹ 4. Regression Diagnostics
Residual plots help evaluate model accuracy and error distribution.

```python
sns.residplot(data=tips, x='total_bill', y='tip')
plt.title('Residual Plot')
plt.show()
```

## ðŸ”¹ 5. Advanced Modeling with lmplot()
Use `lmplot()` for grouped regression models with hue, row, and column faceting.

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker')
plt.title('LMPlot: Regression by Smoker Category')
plt.show()
```
