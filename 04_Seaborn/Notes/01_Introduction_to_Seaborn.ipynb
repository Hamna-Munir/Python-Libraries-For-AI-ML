## What is Seaborn?
Seaborn is a powerful **data visualization library** built on top of Matplotlib. 

It provides:
- High-level, easy-to-use plotting functions
- Beautiful default themes
- Enhanced statistical visualizations
- Built-in datasets

Seaborn helps create graphs with less code and more meaningful visuals.

## Installation
Install Seaborn using:

```bash
pip install seaborn
```

Or in Anaconda:

```bash
conda install seaborn
```

## Difference between Matplotlib & Seaborn

| Feature | Matplotlib | Seaborn |
|--------|------------|---------|
| Abstraction Level | Low (detailed control) | High (simple, statistical) |
| Visual Style | Basic | Beautiful themes |
| Built-in datasets | No | Yes |
| Best For | Custom plots | Statistical plots, EDA |

**Summary:** Matplotlib gives control; Seaborn gives beauty + simplicity.

## Importing Libraries

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
````

## Loading Built-in Datasets
Seaborn provides useful sample datasets to practice on.

Some common datasets:
- `tips`
- `titanic`

Let's load and display a few rows:

````python
tips = sns.load_dataset('tips')
tips.head()
````

## Basic Plot Structure in Seaborn
Seaborn plots follow a very simple structure:

```python
sns.function_name(data=df, x='col1', y='col2')
plt.show()
```

Let's create a simple scatter plot:

````python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Basic Scatterplot')
plt.show()
````

## Axes-level vs Figure-level Functions

Seaborn has two types of plotting functions:

### **1. Axes-Level Functions**
These draw on a single matplotlib axis.
- Examples: `scatterplot`, `lineplot`, `histplot`, `boxplot`
- Return: `matplotlib Axes`

Used when you want **custom layout** or multiple subplots.

### **2. Figure-Level Functions**
These create the entire figure automatically.
- Examples: `relplot`, `catplot`, `lmplot`, `jointplot`
- Return: `FacetGrid` or `Figure`

Used when you need **multiple subplots or faceting**.

````python
sns.relplot(data=tips, x='total_bill', y='tip', col='time')
````
