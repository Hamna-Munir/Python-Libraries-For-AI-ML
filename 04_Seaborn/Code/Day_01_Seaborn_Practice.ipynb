# Seaborn
## 1-Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis
  
Plots under this section  

- scatterplot
- lineplot


```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# Scatter plot -> Axes level function
sns.scatterplot(data = tips , x = 'total_bill' ,y = 'tip', hue='sex' , style='time',size='size')
```

```python
# relplot -> figure level -> square shape
sns.relplot(data = tips , x = 'total_bill' ,y = 'tip', kind = 'scatter', hue='sex' , style='time',size='size')
```

# Line plot

```python
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'Pakistan']
temp_df
```

```python
# Axes level function
sns.lineplot(data=temp_df , x = 'year' , y = 'lifeExp')
```

```python
# using relplot
sns.relplot(data = temp_df , x = 'year' ,y = 'lifeExp', kind = 'line')
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['Brazil','Pakistan','Germany'])]
temp_df
```

```python
sns.relplot(data = temp_df , x = 'year' ,y = 'lifeExp', kind = 'line',hue='country',style='continent',size = 'continent' )
```

```python
# facet plot
sns.relplot(data = tips , x = 'total_bill' ,y = 'tip', kind = 'scatter',col = 'smoker' , row='sex')
```

```python
# facet plot -> figure level function -> work with relplot
# it will not work with scatter plot or line plot
sns.relplot(data = tips , x = 'total_bill' ,y = 'tip', kind = 'line',col = 'smoker' , row='sex')
```

```python
# col wrap 
sns.relplot(data=gap, x = 'lifeExp' , y = 'gdpPercap', kind ='scatter' , col='year' , col_wrap = 3)
```
