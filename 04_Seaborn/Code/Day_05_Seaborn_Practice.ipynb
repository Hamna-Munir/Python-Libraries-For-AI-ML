# *Seaborn*

# Regression Plot

- regplot
- Implot

In the simplest invocation, both functions draw a scatterplot of two variables x and y , and then fit the regression model  y ~ x and 
plot the resulting regression line and a 95% confidence interval for that regression.

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
# regplot -> Axes level function
# hue parameter is not available
sns.regplot(data=tips , x = 'total_bill' , y= 'tip')
```

```python
# lmplot & using hue -> graph level function
sns.lmplot(data=tips , x = 'total_bill' , y= 'tip' , hue = 'sex')
```

# Residualplot

```python
# residplot
sns.residplot(data=tips , x = 'total_bill' , y= 'tip')
```
