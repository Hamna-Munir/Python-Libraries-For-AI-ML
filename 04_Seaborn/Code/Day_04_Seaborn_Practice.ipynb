# Seaborn

# Categorical Plots 

Categorical Scatter Plot

- strip plot
- swarm plot

Categorical Distribution Plots

- Boxplot
- Violinplot

Categorical Estimate Plot -> for central tendency

- Barplot
- Pointplot
- Countplot

Figure level function -> catplot

```python
import seaborn as sns
```

```python
# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

## Strip plot

```python
# strip plot -> Axes level function
sns.stripplot(data=tips , x='day' , y='total_bill')
```

```python
# catplot -> figure level function
sns.catplot(data=tips , x='day' , y='total_bill',kind='strip')
```

```python
# using jitter -> if you give "false" value of jitter than the scatterplot look like this
sns.stripplot(data=tips , x='day' , y='total_bill',jitter = False )
```

```python
# using jitter -> if you give some value of jitter than the scatterplot look like this
sns.stripplot(data=tips , x='day' , y='total_bill',jitter = 0.2 )
```

## Swarm plot

```python
# swarmplot
sns.catplot(data=tips , x='day' , y='total_bill',kind='swarm')
```

```python
# hue
sns.swarmplot(data=tips , x='day' , y='total_bill',hue='sex')
```

# Box plot

```python
# Box plot => Axes level function
sns.boxplot(data= tips , x = 'day' , y = 'total_bill')
```

```python
# using catplot & hue
sns.catplot(data= tips , x = 'day' , y = 'total_bill' , kind = 'box' , hue = 'sex')
```

```python
# single boxplot -> numerical col
sns.boxplot(data=tips , y = 'total_bill')
```

# Violinplot = (Boxplot + KDEplot)

```python
# violinplot -> Axes level function
sns.violinplot(data= tips , x = 'day' , y = 'total_bill' )
```

```python
# using catplot & hue
sns.catplot(data= tips , x = 'day' , y = 'total_bill' , kind = 'violin' , hue = 'sex' , split = True)
```

# Barplot

```python
# Barplot
sns.barplot(data=tips , x = 'sex' , y = 'total_bill' , hue = 'smoker')
```

```python
sns.barplot(data=tips , x = 'sex' , y = 'total_bill' )
```

# Pointplot

```python
# Pointplot
sns.pointplot(data=tips , x = 'sex' , y = 'total_bill')
```

```python
# using hue 
sns.pointplot(data=tips , x = 'sex' , y = 'total_bill' , hue = 'smoker')
```

# Countplot

```python
# countplot
sns.countplot(data=tips , x = 'sex')
```

```python
# countplot using hue
sns.countplot(data=tips , x = 'sex' , hue = 'day')
```

# Faceting using plot


```python
# faceting plot 
sns.catplot(data= tips , x = 'sex' , y = 'total_bill' , kind = 'box' , col = 'smoker' , row = 'time')
```
