# *Seaborn*

# Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data Bimodal?
- Are there outliers?

Plots under distribution plot

- Histplot
- Kdeplot
- Rugplot


```python
# figure level -> displot
# Axes level -> histplot -> kdeplot -> rugplot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# Histplot

```python
# plotting Univartiate histogram
sns.histplot(data = tips , x = 'total_bill')
```

```python
sns.displot(data=tips , x = 'total_bill' , kind='hist')
```

```python
# its also possible to visulize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables 

# countplot 
sns.displot(data=tips, x = 'day' , kind = 'hist')
```

```python
# hue parameter
sns.displot(data=tips, x = 'day' , kind = 'hist', hue = 'sex')
```

```python
# element -> step
sns.displot(data=tips, x = 'day' , kind = 'hist' , hue = 'sex' , element = 'step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x = 'age' , kind = 'hist' , hue = 'sex' , element = 'step')
```

```python
# faceting using row and column -> not work on histplot function
sns.displot(data=tips, x = 'tip' , kind = 'hist' , col='sex' , element='step')
```

# Kdeplot

```python
# kdeplot
# Rather than using discerete bins , a KDE Plot smooths the observations with a Gussian kernal , producing a countinuous density estimate
sns.kdeplot(data=tips , x = 'total_bill')
```

```python
# use displot
sns.displot(data=tips , x = 'total_bill', kind = 'kde')
```

```python
# hue -> fill
sns.displot(data=tips , x = 'total_bill', kind = 'kde', hue = 'sex',fill = True)
```

# Rugplot

```python
# plot marginal distributions by drawing ticks along the x and y axis 

# this function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips , x = 'total_bill')
sns.rugplot(data=tips , x = 'total_bill')
```

# Bivariate Histogram

```python
# A bivariate histogram bins the data within rectangles that title the plot 
# and then shows the count of observations within each rectangle with the fill colour 

sns.histplot(data=tips , x = 'total_bill' , y='tip')
```

# Bivariate kdeplot


```python
# A bivariate KDE plot smoothes the (x,y) observations with a 2D Guassian

sns.kdeplot(data=tips , x = 'total_bill' , y = 'tip')
```
