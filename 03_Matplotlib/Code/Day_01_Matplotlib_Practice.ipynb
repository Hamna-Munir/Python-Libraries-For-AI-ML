```python
# import the libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

plt.style.use("classic")
```

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman

plt.plot(batsman['index'] ,batsman['V Kohli'])
```

```python
# labels title
plt.plot(batsman['index'] ,batsman['V Kohli'])
plt.plot(batsman['index'] ,batsman['RG Sharma'])

plt.title("Rohit Sharma Vs Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
#what is the mistake in code "# colours(hex) and line (width and style) and marker (size)
plt.plot(batsman['index'], batsman['V Kohli'], color='#F54927')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#170501')

plt.title("Rohit Sharma Vs Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'],linestyle='dashed')
plt.plot(batsman['index'], batsman['RG Sharma'],linestyle='dashed')

plt.title("Rohit Sharma Vs Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'],color='#941601',linestyle='dotted')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#121010',linestyle='dotted')

plt.title("Rohit Sharma Vs Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#CC7722', linestyle='dashdot')
plt.plot(batsman['index'], batsman['V Kohli'], color='#A52A2A', linestyle='dashdot')
plt.plot(batsman['index'], batsman['RG Sharma'], linestyle='dashdot')

plt.title("Rohit Sharma Vs Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#8B0000', linestyle='solid',linewidth=3)
plt.plot(batsman['index'], batsman['V Kohli'], color='#483C32', linestyle='dashdot',linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'], linestyle='dashdot')

plt.title("Rohit Sharma Vs Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
# use marker and markersize
plt.plot(batsman['index'], batsman['V Kohli'], color='#8B0000', linestyle='solid',linewidth=3,marker='D')
plt.plot(batsman['index'], batsman['V Kohli'], color='#483C32', linestyle='dashdot',linewidth=3,marker='+')
plt.plot(batsman['index'], batsman['RG Sharma'], linestyle='dashdot')

plt.title("Rohit Sharma Vs Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
# use grid
plt.plot(batsman['index'], batsman['V Kohli'], color='#8B0000', linestyle='solid',linewidth=3,marker='D')
plt.plot(batsman['index'], batsman['V Kohli'], color='#483C32', linestyle='dashdot',linewidth=3,marker='+')
plt.plot(batsman['index'], batsman['RG Sharma'], linestyle='dashdot')

plt.title("Rohit Sharma Vs Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.grid()
plt.show
```
