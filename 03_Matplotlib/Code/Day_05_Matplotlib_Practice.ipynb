# Pie Chart
- Univariate/Bivariate Analysis
- Categorical VS Numerical
- Use case - To find contribution on a random scale

```python
# import the libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Simple Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Title
plt.title('Programming Language Usage')

# Display chart
plt.show()
```

```python
# Use dataset
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%' )
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%', explode=[0.2,0,0,0,0,0],shadow=True )
plt.show()
```
