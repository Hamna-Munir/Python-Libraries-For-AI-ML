# Colored Scatterplots

```python
# import the libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],alpha=0.7)
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.colorbar()
plt.show()
```

# Plot size

```python
# Increase the size of graph
plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],alpha=0.7)
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.colorbar()
plt.show()
```

# Anotations

```python
batters = pd.read_csv('batter.csv')
batters.head()
```

```python
sample_batters = batters.head(100).sample(25,random_state=5)
sample_batters
```

```python
# Use Anotations
plt.figure(figsize=(18,10))
plt.scatter(sample_batters['avg'],sample_batters['strike_rate'],s=sample_batters['runs'])

plt.axhline(130,color='red')
plt.axvline(30,color='red')

for i in range(sample_batters.shape[0]):
    plt.text(sample_batters['avg'].values[i],sample_batters['strike_rate'].values[i],sample_batters['batter'].values[i])
    
```
