```python
# import the libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# simple bar chart 
children = [10 , 5 , 25 , 15 , 30]
colour = ['red' , 'blue' , 'green' , 'pink' , 'yellow']
plt.bar(colour, children, color='#3498DB')
```

```python
# Horizontal bar chart 
plt.barh(colour, children, color='#3498DB')
```

```python
# clor and label
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# we use a "plt.xticks(rotation='vertical')" if the categories name in X-Axis is lenghty then this function print name vertically and enhance readibility
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Stacked bar chart
# Data
categories = ['A', 'B', 'C', 'D']
boys = [20, 35, 30, 35]
girls = [25, 32, 34, 20]

# Create stacked bar chart
plt.bar(categories, boys, label='Boys')
plt.bar(categories, girls, bottom=boys, label='Girls')

# Labels and title
plt.xlabel('Categories')
plt.ylabel('Number of Students')
plt.title('Stacked Bar Chart Example')
plt.legend()

# Show plot
plt.show()
```
