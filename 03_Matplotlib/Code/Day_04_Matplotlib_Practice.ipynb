# Histogram
- Univariate Analysis
- Numerical Col
- Use case - Frequency count

```python
# import the libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#simple data
data = [10, 20, 20, 30, 30, 30, 40, 50]

plt.hist(data, bins=4, edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')

plt.show()
```

# Logarithmic Scale is Used for Solving These type of problems

- Problems with Very Large Data Ranges
- Exponential Growth or Decay Problems
- Algorithm & Time Complexity Problems (CS / AI)
- Power-Law Distribution Problems
- Error, Loss, and Convergence Problems (AI / ML)
## Log scale makes small and large values visible together.

```python
# logarithmic scale
arr = np.load('big-array.npy')
plt.hist(arr,bins=[10,20,30,40,50,60,70],log=True)
plt.show()
```
