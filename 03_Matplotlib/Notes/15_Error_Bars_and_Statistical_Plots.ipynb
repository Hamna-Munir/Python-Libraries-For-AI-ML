## Introduction
In data analysis, it is essential not only to visualize raw values but also to represent the **uncertainty, variation, and statistical distribution** of data. Matplotlib provides a rich set of tools to visualize statistical information.

In this notebook, we will cover:
- What error bars are and why they matter

- Representing statistical variation using boxplots

- Creating histograms and density plots

Mastering these techniques is crucial for scientific, academic, and analytical work involving uncertainty and statistical data.

# 1. Understanding Error Bars
Error bars show the **uncertainty or variability** in a dataset. They help readers understand the reliability of the measured values.

Error bars can represent:
- Standard deviation
- Standard error
- Measurement error
- Confidence intervals

Matplotlib provides `plt.errorbar()` to create error bar plots.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y = np.array([10, 12, 8, 15, 14])
errors = np.array([1.5, 2.0, 1.0, 2.5, 1.8])

plt.errorbar(x, y, yerr=errors, fmt='o-', capsize=5)
plt.title("Simple Error Bar Plot")
plt.xlabel("Sample")
plt.ylabel("Value")
plt.show()
```

# 2. Adding Horizontal and Vertical Error Bars
Matplotlib supports both **x-direction** and **y-direction** error bars.

- `yerr` → vertical error
- `xerr` → horizontal error
- `capsize` → cap size at the ends of error bars
- `fmt` → marker + line style

```python
xerr = np.array([0.2, 0.3, 0.1, 0.25, 0.2])

plt.errorbar(x, y, xerr=xerr, yerr=errors,
             fmt='s--', capsize=6, ecolor='gray')

plt.title("Horizontal and Vertical Error Bars")
plt.xlabel("Sample")
plt.ylabel("Value")
plt.show()
```

# 3. Error Bars with Custom Colors and Line Styles
Error bars can be styled for clarity and presentation.

You can customize:
- `color` → line + marker color
- `ecolor` → error bar color
- `elinewidth` → thickness of error line
- `capsize` → cap width
- `linestyle` → line between points


```python
plt.errorbar(x, y, yerr=errors,
             fmt='o-', color='purple', ecolor='black',
             elinewidth=2, capsize=7, linestyle='--')

plt.title("Styled Error Bars")
plt.xlabel("Sample")
plt.ylabel("Measurement")
plt.show()
```

# 4. Boxplots – Visualizing Statistical Distribution
Boxplots summarize data distribution using five key metrics:
- Minimum
- First quartile (Q1)
- Median (Q2)
- Third quartile (Q3)
- Maximum

They are excellent for spotting:
- Variability
- Skewness
- Outliers
- Spread of data


```python
data = np.random.normal(50, 10, 200)

plt.boxplot(data, vert=True, patch_artist=True)
plt.title("Boxplot – Statistical Distribution")
plt.ylabel("Values")
plt.show()
```

# 5. Multiple Boxplots for Comparison
Multiple datasets can be compared easily using grouped boxplots.

```python
group1 = np.random.normal(60, 12, 200)
group2 = np.random.normal(55, 15, 200)
group3 = np.random.normal(50, 8, 200)

plt.boxplot([group1, group2, group3], patch_artist=True)
plt.title("Multiple Boxplots – Group Comparison")
plt.xlabel("Group")
plt.ylabel("Values")
plt.show()
```

# 6. Violin Plots – Understanding Data Shape
A violin plot shows the **distribution of the data**, combining a boxplot with a density curve.

It is useful for:
- Seeing multimodal distributions
- Understanding the shape of the data
- Comparing statistical patterns across groups

```python
plt.violinplot([group1, group2, group3], showmeans=True)
plt.title("Violin Plot – Density + Statistics")
plt.xlabel("Group")
plt.ylabel("Values")
plt.show()
```

# 7. Histograms – Frequency Distribution
Histograms summarize how often values fall into specific ranges.

Useful for:
- Detecting skewness
- Identifying outliers
- Observing spread and distribution pattern
- Understanding frequency of values


```python
plt.hist(data, bins=20, edgecolor='black')
plt.title("Histogram – Frequency Distribution")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

# 8. Density Plot (PDF Curve)
Density plots show a smooth probability distribution curve.

They are created using `plt.plot()` over sorted values and probability density.

```python
from scipy.stats import gaussian_kde

density = gaussian_kde(data)
x_vals = np.linspace(min(data), max(data), 200)

plt.plot(x_vals, density(x_vals))
plt.title("Density Curve (PDF)")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()
```
