## 1. Importing Required Libraries
We begin by importing `matplotlib.pyplot` for plotting and enabling inline display inside Jupyter Notebook.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

---
## 2. Line Plot
A line plot is one of the most commonly used charts to show trends over time or continuous data.  
    It connects data points using straight lines.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

---
## 3. Scatter Plot
Scatter plots are useful for showing relationships between two variables.  
    Each point represents an observation in the dataset.

```python
x = [5, 7, 8, 10, 12]
y = [12, 16, 20, 22, 28]

plt.figure(figsize=(6,4))
plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

---
## 4. Bar Chart
Bar charts are ideal for comparing categories or groups.

```python
categories = ["A", "B", "C", "D"]
values = [10, 24, 36, 18]

plt.figure(figsize=(6,4))
plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

---
## 5. Histogram
Histograms show the distribution of numerical data by grouping values into bins.

```python
data = [12, 14, 15, 19, 22, 25, 25, 28, 30, 32, 35, 38]

plt.figure(figsize=(6,4))
plt.hist(data, bins=5)
plt.title("Histogram Example")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

---
## 6. Pie Chart
Pie charts show percentage distribution across categories.

```python
sizes = [40, 25, 20, 15]
labels = ["Category A", "Category B", "Category C", "Category D"]

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```
