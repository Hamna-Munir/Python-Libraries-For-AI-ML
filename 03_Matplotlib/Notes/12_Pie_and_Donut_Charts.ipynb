## Introduction
Pie charts are useful for showing **proportions or percentage distribution** in a dataset.
Donut charts are an extension of pie charts with a **hole in the center**, making them more visually appealing.

In this notebook, you will learn:
- How to create pie charts
- How to customize them (labels, colors, explode, startangle)
- How to convert pie charts into donut charts
- How to add annotations and percentages


# 1. Creating a Basic Pie Chart
A pie chart represents each category as a slice of the whole.

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 25, 25]

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Basic Pie Chart')
plt.show()
```

# 2. Customizing Pie Chart
You can customize using:
- `colors`
- `explode` to pull out slices
- `startangle` to rotate chart


```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = [0.1, 0, 0, 0]  # highlight 1st slice

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, explode=explode,
        autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Customized Pie Chart')
plt.show()
```

# 3. Creating a Donut Chart
A donut chart is simply a pie chart with a white circle in the center.

```python
plt.figure(figsize=(7,7))
wedges, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Create donut hole
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Donut Chart')
plt.show()
```

# 4. Donut Chart with Multiple Rings
You can create multiple layers to represent hierarchical data.

```python
inner_sizes = [15, 25, 35, 25]
outer_sizes = [30, 20, 25, 25]

fig, ax = plt.subplots(figsize=(8,8))
ax.pie(outer_sizes, radius=1.2, labels=labels, autopct='%1.1f%%')
ax.pie(inner_sizes, radius=0.8, autopct='%1.1f%%')

# Add center
centre = plt.Circle((0,0), 0.5, color='white')
ax.add_artist(centre)

plt.title('Multi-Ring Donut Chart')
plt.show()
```

# Summary
- Pie charts represent proportional data.
- Customization options include colors, explode, angles, shadows.
- Donut charts are pie charts with a center hole, improving visual appeal.
- Multi-ring donuts help visualize layered or hierarchical data.

These charts are great for business dashboards, category-based distributions, and reporting.
