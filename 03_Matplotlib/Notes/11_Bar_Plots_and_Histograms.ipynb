## Introduction
Bar plots are used to visualize **categorical data**, while histograms visualize the **distribution of numerical data**.


## 1. Bar Plots
A bar plot displays rectangular bars whose heights represent values.

### Important Parameters
- `x`: category labels
- `height`: values
- `color`: bar color
- `width`: bar width
- `align`: center or edge alignment


```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 25]

plt.bar(categories, values)
plt.title("Basic Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

## 2. Horizontal Bar Plot
Useful when category text is long or when horizontal comparison is more readable.


```python
plt.barh(categories, values)
plt.title("Horizontal Bar Plot")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```

## 3. Stacked Bar Chart
Stacked bars show the total value along with category-wise contribution.


```python
group1 = [10, 15, 20, 25]
group2 = [5, 10, 15, 10]

plt.bar(categories, group1)
plt.bar(categories, group2, bottom=group1)
plt.title("Stacked Bar Chart")
plt.show()
```

# Histograms
Histograms display **how data is distributed** across ranges.

### When to Use a Histogram?
- Understanding spread of data
- Finding skewness
- Detecting outliers
- Observing shape of distribution

### Key Parameter: `bins`
- Number of intervals
- More bins → detailed distribution
- Fewer bins → generalized overview


```python
import numpy as np

data = np.random.randn(200)

plt.hist(data, bins=20)
plt.title("Histogram Example")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```
