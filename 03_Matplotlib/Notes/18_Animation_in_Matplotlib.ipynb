## Introduction
Matplotlib supports animations through the **`matplotlib.animation`** module.

Animations are useful for:
- Time-series visualizations
- Dynamic simulations
- Educational demonstrations
- Updating plots in real time

In this notebook, we will cover:
- Line animation
- Scatter point animation
- Using `FuncAnimation`
- Displaying animations inside Jupyter Notebook


# 1. Importing Required Libraries

````python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
````

# 2. Line Animation
This animation updates a sine wave over time.

````python
fig, ax = plt.subplots()
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.2, 1.2)

line, = ax.plot([], [], lw=2)

def init_line():
    line.set_data([], [])
    return line,

def animate_line(i):
    x = np.linspace(0, 2*np.pi, 200)
    y = np.sin(x + 0.1*i)
    line.set_data(x, y)
    return line,

anim_line = FuncAnimation(fig, animate_line, init_func=init_line,
                          frames=100, interval=50, blit=True)

HTML(anim_line.to_jshtml())
````

# 3. Scatter Animation
This animation adds random scatter points over time.

````python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

fig2, ax2 = plt.subplots()
ax2.set_xlim(0, 10)
ax2.set_ylim(0, 10)

scatter = ax2.scatter([], [])
xdata, ydata = [], []

def init_scatter():
    scatter.set_offsets(np.empty((0, 2)))   # FIX
    return scatter,

def animate_scatter(i):
    xdata.append(np.random.rand()*10)
    ydata.append(np.random.rand()*10)
    points = np.column_stack((xdata, ydata))
    scatter.set_offsets(points)
    return scatter,

anim_scatter = FuncAnimation(
    fig2, animate_scatter, init_func=init_scatter,
    frames=50, interval=200, blit=True
)

HTML(anim_scatter.to_jshtml())
````

# 4. Saving Animations
You can save animations as **MP4** or **GIF** (requires ffmpeg or imagemagick):

```python
anim_line.save('line_animation.mp4', writer='ffmpeg')
anim_scatter.save('scatter.gif', writer='imagemagick')
```

# Summary
- `FuncAnimation` creates dynamic plots.
- Line animations show changes in continuous functions.
- Scatter animations update point positions over time.
- `HTML(anim.to_jshtml())` is used to display animations inside Jupyter.
- Animations can be exported as videos or GIFs.
