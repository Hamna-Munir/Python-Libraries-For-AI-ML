# List Vs Arrays
* Faster
* Convenient
* Less Memory

```python
import numpy as np
```

```python
# Array occupy less memory as compare to list

lista = range(100)
arr1 = np.arange(100)
```

```python
import sys
```

```python
print(sys.getsizeof(87)*len(lista))
```

```python
print(arr1.itemsize*arr1.size)
```

```python
# Array is faster than list
import time
```

```python
# List
x = range(10000000)
y = range(10000000,20000000)

start_time = time.time()
c = [x+y for x,y in zip(x,y)]

print(time.time()-start_time)
```

```python
# Array
a = np.arange(10000000)
b = np.arange(10000000,20000000)

start_time = time.time()
c = a+b

print(time.time()-start_time)
```
