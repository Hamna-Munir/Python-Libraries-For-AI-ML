```python
# Replace the odd values of the array with "0"

a = np.random.randint(1,10,6)
a
```

```python
# Use the "Where" functions to replace the values of the array

a = np.random.randint(1,50,6)
a
```

```python
np.where(a%2==1 , -1 ,a)
```
